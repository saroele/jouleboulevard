```python
import pandas as pd
import json
```

```python
df = pd.read_csv('Jouleboulevard_metadata.csv', index_col='MeterId', usecols=['MeterId', 'Settings'])
```

```python
def parse_settings(x):
    j = json.loads(x)
    return j['sensor'], j['token']
```

```python
df['SensorId'], df['Token'] = df.Settings.map(parse_settings).str
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(columns=['MeterId', 'Settings'], inplace=True)
```

```python
df.to_csv('jouleboulevard_fluksos.csv')
```
