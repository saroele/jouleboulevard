```python
import pandas as pd
import json
```

```python
df = pd.read_csv('Jouleboulevard_metadata_raw.csv', index_col='MeterId')
```

```python
def parse_settings(x):
    j = json.loads(x)
    return j['sensor'], j['token']
```

```python
df['SensorId'], df['Token'] = df.Settings.map(parse_settings).str
```

```python
df.drop(columns=['Settings'], inplace=True)
```

```python
df.to_csv('jouleboulevard_metadata.csv')
```
